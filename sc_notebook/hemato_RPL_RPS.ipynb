```python
df = pd.DataFrame(columns = sonar['gene name'])

for clus in adata.uns['rank_genes_groups']['names'].dtype.names:
    gene = adata.uns['rank_genes_groups']['names'][clus]
    #zscore = adata.uns['rank_genes_groups']['scores'][clus]
    rbp = set(gene).intersection(set(sonar['gene name']))
    df.loc[clus, list(rbp)] = True
df.fillna(False, inplace = True)
df = df.loc[:, df.sum(axis = 0)>0]
    
    
    
    
```

```python
import seaborn as sns
sns.clustermap(df.loc[:, df.columns[(df.columns.str.contains('RPS')) | df.columns.str.contains('RPL')]]
               , xticklabels=1
              , figsize=(20, 3))
```
