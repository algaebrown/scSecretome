```r
library(ReactomePA)
library(repr)
parse_genelist <- function(file){
    # from de gene list to r list
    t = read.csv(file = file, sep = '\t', header = FALSE)
    
    # make into R list
    t$V2 = sapply(t$V2, toString)
    l = sapply(t$V2, strsplit, ',')
    names(l) = t$V1
    
    return(l)
    }
```

# Reactome enrichment analysis on differentially expressed genes
## Haematopoietic stem cells

```r
pellin_cd34 = parse_genelist('pellin_34_de')
```

```r
reactome_enrichment <- function(gene_list){
    de = as.integer(as.vector(unlist(gene_list)))
    
    # run enrichment analysis
    x <- enrichPathway(gene=de,pvalueCutoff=0.05, readable=T)
    
    return(x)
}
```

```r
x <- reactome_enrichment(pellin_cd34['8'])
```

```r
# Change plot size to 4 x 3
options(repr.plot.width=10, repr.plot.height=5)

barplot(x, showCategory=20, font.size = 5)
```

```r
cnetplot(x)
```

```r
emapplot(x)
```

```r
require(clusterProfiler)
res <- compareCluster(pellin_cd34, fun="enrichPathway")
```
