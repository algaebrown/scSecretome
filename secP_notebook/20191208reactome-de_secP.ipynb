```r
library(ReactomePA)
library(repr)
require(clusterProfiler)
parse_genelist <- function(file){
    # from de gene list to r list
    t = read.csv(file = file, sep = '\t', header = FALSE)
    
    # make into R list
    t$V2 = sapply(t$V2, toString)
    l = sapply(t$V2, strsplit, ',')
    names(l) = t$V1
    
    return(l)
    }
reactome_enrichment <- function(gene_list){
    # reactome enrichment analysis for a single cluster: ex: gene_list['8']
    de = as.integer(as.vector(unlist(gene_list)))
    
    # run enrichment analysis
    x <- enrichPathway(gene=de,pvalueCutoff=0.05, readable=T)
    
    return(x)
}
```

# Reactome enrichment analysis on differentially expressed SECRETORY PROTEIN

```r
pellin_cd34 = parse_genelist('pellin_cd34_sec_de')
res <- compareCluster(pellin_cd34, fun="enrichPathway")

```
