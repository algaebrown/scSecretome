```r
library(ReactomePA)
```

```r
path <- "cell12_symbol.txt"
print(path)
 
conn <- file(path,open="r")
lines <- readLines(conn)

close(conn)


```

```r
path <- "cell12_symbol2.txt"
print(path)
 
conn <- file(path,open="r")
lines2 <- readLines(conn)

close(conn)


```

```r
library(rlist)
parse_genelist <- function(file_lines){
    gene_list <- list()
    cell_id = 0
    for (i in length(file_lines)){
        gene_list[cell_id] <- strsplit(lines[cell_id+1], ',')
        
        cell_id = cell_id + 1
        }
    return(gene_list)
}
```

```r
gene_list = sapply(lines, strsplit, split = ',')
names(gene_list) <- seq(0, length(lines)-1)
```

```r
gene_list2 = sapply(lines2, strsplit, split = ',')
names(gene_list2) <- seq(0, length(lines)-1)
```

```r
reactome_enrichment <- function(cell_id){
    de = as.integer(as.vector(unlist(strsplit(lines[cell_id+1], ','))))
    
    # run enrichment analysis
    x <- enrichPathway(gene=de,pvalueCutoff=0.05, readable=T)
    
    return(x)
}
```

```r
x <- reactome_enrichment(0)
barplot(x, showCategory=20)
```

```r
cnetplot(x)
```

```r
emapplot(x)
```
